```python
# Adapted from https://www.freecodecamp.org/news/how-to-combine-multiple-csv-files-with-8-lines-of-code-265183e0854/
# Consider loading in chunks if memory limitations ocurr. https://stackoverflow.com/questions/56494140/how-do-i-combine-large-csv-files-in-python

import os
import glob
import pandas as pd
os.chdir("CSV_combined/")

extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]

#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f, names=['posX', 'posY', 'posZ', 'A', 'B', 'C', 'Z']) for f in all_filenames ])

# remove data points outside mapping area
width = 900
hWidth = int(width/2)
combined_csv = combined_csv.loc[(combined_csv['posX'] >= -hWidth) & (combined_csv['posX'] <= hWidth)]

#export to csv
combined_csv.to_csv( "combined.csv", header=False, index=False, encoding='utf-8-sig')
os.chdir("..")
```
