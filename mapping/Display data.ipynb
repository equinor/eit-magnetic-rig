```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

filename = 'combined.csv'

df_combined = pd.read_csv('CSV_combined/' + filename, names=['posX', 'posY', 'posZ', 'A', 'B', 'C', 'Z'])
df_combined
```

```python
plt.figure(figsize=(16, 1), dpi=(200))
plt.scatter(df_combined['posX'], df_combined['posY'], c=df_combined['C'], s=0.1, edgecolors='none', cmap = 'viridis', vmin=0, vmax=35000)
plt.savefig('combined.png', dpi=2400)
```

```python
df_sliceZ = df_combined.loc[(df_combined['posZ'] == 155)]
df_sliceZ = df_sliceZ.round(decimals=1)
df_sliceZ.drop_duplicates(subset=['posX', 'posY', 'posZ'], keep='last',inplace=True)
    
plt.figure(figsize=(9, 6), dpi=(100))
plt.scatter(df_sliceZ['posX'], df_sliceZ['posY'], c=df_sliceZ['C'], marker='s', s=4, edgecolors='none', cmap = 'viridis', vmin=0, vmax=35000)
plt.savefig('sliceZ.png', dpi=200)

df_sliceZ
```

```python
df_sliceX = df_combined.loc[(df_combined['posY'] == 0)]
df_sliceX = df_sliceX.round(decimals=1)
df_sliceX.drop_duplicates(subset=['posX', 'posY', 'posZ'], keep='last',inplace=True)
    
plt.figure(figsize=(9, 6), dpi=(100))
plt.scatter(df_sliceX['posX'], df_sliceX['posZ'], c=df_sliceX['C'], marker='s', s=4, edgecolors='none', cmap = 'viridis', vmin=0, vmax=35000)
plt.savefig('sliceX.png', dpi=200)

df_sliceY
```
